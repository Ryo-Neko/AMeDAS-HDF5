# Example of usage for AMeDAS HDF5 dataset

## Data Description

### Structure of HDF5
You do not need to consider about the structure during you use the Class below.

Structure is:
- /data: data
- / qualities: The quality flag fllowing JMA method (https://www.data.jma.go.jp/obd/stats/data/kaisetu/shishin/shishin_3.pdf). Converted as a flag like below tables. 

After the two flags: /climatic_element/year/MonthDate

# Example of usage

```python
import warnings

import h5py as h5
import pandas as pd
import numpy as np

import datetime as dt
```

```python
class AMeDASDump:
    """Class for retrieving and dumping AMeDAS data for specified period and element at specified stations.

    Attributes:
        start_date (str): Start date for the data retrieval period in YYYY-MM-DD format.
        end_date (str): End date for the data retrieval period in YYYY-MM-DD format.
        data_path (str): Path to the directory where the data files are stored.
        climatic_elements (list): List of climatic elements to retrieve data for. Default is ["Precipitation"].
        dataset (pandas.DataFrame): DataFrame containing the retrieved data.
        stations (list): List of station IDs and names for the stations to retrieve data for.
        ame_master (pandas.DataFrame): DataFrame containing the AMeDAS station master data.

    Methods:
        make_station_list(station_ids: list) -> list: Make a list of station IDs and names for the stations to retrieve data for.
        dumpler(station_ids: list) -> None: Dump data for specified period and element at specified stations.
        reset() -> None: Reset the dataset and stations attributes to empty list and DataFrame, respectively.
    """
    def __init__(self, start_date, end_date, data_path, climatic_elements=["Precipitation"]):
        
        self.start_date = start_date
        self.end_date = end_date
        self.data_path = data_path
        self.climatic_elements = climatic_elements

        self.__ame_master = pd.read_csv("ame_master_le.csv", header=0, index_col=0, encoding="utf-8")
        self.__dataset = pd.DataFrame()
        self.__stations = []

    @property
    def dataset(self):
        return self.__dataset
    
    @property
    def stations(self):
        return self.__stations
    
    @property
    def ame_master(self):
        return self.__ame_master

    def reset(self):
        self.__dataset = pd.DataFrame()
        self.__stations = []

    def make_station_list(self, station_ids: list) -> list:
        if station_ids:
            df_ = self.__ame_master.loc[self.__ame_master["1"].isin(station_ids)]
            self.__stations = [id_ + "_" + name for id_, name in zip(df_["1"].apply(lambda x: str(x).zfill(4)).values, df_["2"].values)]
        else:
            self.__stations = [id_ + "_" + name for id_, name in zip(self.__ame_master["1"].apply(lambda x: str(x).zfill(4)).values, self.__ame_master["2"].values)]
        if ((self.__stations == []) or (self.__stations is None)):
            raise ValueError("All station_ids are not valid.")
        return 

    def dumpler(self, station_ids: list) -> None:
        """Retrieve data for specified period and element at specified stations.

        Args:
            station_ids (list): List of IDs for the stations to retrieve data for.
        Returns:
            None
        """
        self.make_station_list(station_ids)
        for station in self.__stations:
            for climatic_element in self.climatic_elements:
                print(station, climatic_element)
                station_id = station.split("_")[0]
                station_name = station.split("_")[1]
                # Path for HDF5
                file_path = f"{self.data_path}/{station_id}.h5"
                with h5.File(file_path, 'r') as hdf_file:
                    # Make date range for the next loop
                    date_range = pd.date_range(start=self.start_date, end=self.end_date, freq='D')
                    # Make date range for the index in self.all_df
                    date_range_hourly = pd.date_range(start=self.start_date, end=self.end_date + " 23:00", freq='H')
                    # List for the data tentatively
                    data_list = []
                    data_quality_list = []
                    # -------------data getting loop-------------------
                    for date in date_range:
                        dataset_path = f'/data/{climatic_element}/{date.year}/{date.strftime("%m%d")}'
                        dataset_quality_path = f'/qualities/{climatic_element}/{date.year}/{date.strftime("%m%d")}'
                        data_available = True
                        if dataset_path in hdf_file:
                            data = hdf_file[dataset_path][:]
                            data_quality = hdf_file[dataset_quality_path][:]
                            data_list.extend(data)
                            data_quality_list.extend(data_quality)
                        else:
                            warnings.simplefilter("always")
                            warnings.warn(f"The dataset path ({dataset_path}) was not found in the dataset of {station}. "
                            + f"ALL of the '{climatic_element}' was IGNORED in this run.")
                            data_available = False
                            break

                    if data_available:
                        merged_data = pd.DataFrame(np.array([data_list, data_quality_list]).T,
                                                index=date_range_hourly,
                                                columns=pd.MultiIndex.from_arrays([[station_id, station_id], [climatic_element, "Quality_" + climatic_element]]))
                        self.__dataset = pd.concat([self.__dataset, merged_data], axis=1)
```

```python
# station ids to be retrieved. List includes station ids as integer values.
station_ids = [47662, 47590, 1060, 40, 788]

# Period to be retrieved
start_date = '2017-07-01'
end_date = '2023-01-31'

# Climatic elements to be retrieved
climatic_elements = ["Global solar radiation", "Temperature", "Precipitation"]

# Path to save the retrieved data
data_path = "./stations"
```

```python
ame = AMeDASDump(start_date, end_date, data_path, climatic_elements=climatic_elements)
ame.dumpler(station_ids)
```
